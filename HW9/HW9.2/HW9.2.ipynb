```python
import pandas as pd
```

```python
geo_data = {
    'Центр': ['москва', 'тула', 'ярославль'],
    'Северо-Запад': ['петербург', 'псков', 'мурманск'],
    'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
```

```python
data = pd.read_csv('keywords.csv', sep= r',', encoding='utf-8')
data.head()
```

```python
def region_from_data(row):
    data = row['keyword'].split(' ')
    for i in data:
        for items in geo_data.items():
            if i in items[1]:
                return items[0]
    return ''       
```

```python
data['region'] = data.apply(region_from_data, axis=1)
data.head()
```
